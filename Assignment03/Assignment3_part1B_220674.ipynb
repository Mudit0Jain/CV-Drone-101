```python
from PIL import Image, ImageDraw
```

```python
import matplotlib.pyplot as plt
import math
```

```python
def drawFlag():
  flag = Image.new("RGB", (600,600), "white")
  draw = ImageDraw.Draw(flag)
  draw.rectangle([(0,0), (600,200)], fill = "#FF9933")
  draw.rectangle([(0,400),(600,600)], fill ="#138808")

  c = (300,300)
  r = 100
  draw.ellipse([(c[0]-r, c[1]- r), (c[0]+r, c[1]+r)], outline = "#000080", width = 2)
  #draw.line([c,(c[0]+r, c[1]+r) ], fill="#000080", width=1)
  for arg in range(0,360,15):
    x2 = c[0] + r*math.cos(math.radians(arg))
    y2 = c[1] + r*math.sin(math.radians(arg))
    second_point = (x2,y2)
    draw.line([c,second_point ], fill="#000080", width=1)
  #plt.imshow(flag)
  return flag
```

```python
Flag = drawFlag()
```

```python
def rotate(img, y):
  img = img.rotate(y)
  return img
```

```python

plt.imshow(Flag)
```

```python
Flag_90 = rotate(Flag, 90)
```

```python
plt.imshow(Flag_90)
```

```python
def rotatedflags(img):
  img_90 = rotate(img,90)
  img_270 = rotate(img,270)
  img_180 = rotate(img,180)

  fig = plt.figure(figsize=(10,7))

  row=2
  col=2

  fig.add_subplot(row,col,1)
  plt.imshow(img)
  plt.axis('off')

  fig.add_subplot(row,col,2)
  plt.imshow(img_90)
  plt.axis('off')

  fig.add_subplot(row,col,3)
  plt.imshow(img_270)
  plt.axis('off')

  fig.add_subplot(row,col,4)
  plt.imshow(img_180)
  plt.axis('off')
```

```python
rotatedflags(Flag)
```
