```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/Computer vision/kakashi simple cartoon.jpg')
plt.imshow(img[:,:,-1])
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray_inv = cv2.bitwise_not(gray)
gray_inv
```

```python
_, binary = cv2.threshold(gray_inv,47,255,cv2.THRESH_BINARY)
plt.imshow(binary,cmap="gray");plt.title("Binary Image");plt.axis("off")
```

```python
contour, hierarchy = cv2.findContours(binary,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
imgcopy = img.copy()
cv2.drawContours(imgcopy,contour,-1,(255,0,0),2)
plt.figure(figsize=[10,10])
plt.imshow(imgcopy[:,:,-1])
plt.title("Original Image")
plt.axis("off")
```

```python
blurred = cv2.GaussianBlur(img,(5,5),0)
Edges = cv2.Canny(blurred,100,160)
plt.figure(figsize=[10,10])
plt.imshow(Edges,cmap="gray");plt.title("Edge Image")
```

```python
contour, hierarchy = cv2.findContours(Edges,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
imgcopy = img.copy()
for i in range(9):
  cv2.drawContours(imgcopy,[contour[-i]],-1,(255,0,0),2)
plt.figure(figsize=[10,10])
plt.imshow(imgcopy[:,:,-1])
plt.title("Original Image")
plt.axis("off")
```
