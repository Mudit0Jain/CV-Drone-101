```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
import cv2
from google.colab.patches import cv2_imshow
```

```python
url = '/content/drive/MyDrive/Computer vision/image for mideval.jpg'
```

```python
im = cv2.imread(url)
```

```python
def adjust_saturation_brightness_contrast(image, saturation_factor=1.5, brightness_factor=1.5, contrast_factor=1.5):
    # Convert the image to the HSV color space
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Adjust saturation
    hsv[:, :, 1] = np.clip(hsv[:, :, 1] * saturation_factor, 0, 255).astype(np.uint8)

    # Adjust brightness
    hsv[:, :, 2] = np.clip(hsv[:, :, 2] * brightness_factor, 0, 255).astype(np.uint8)

    # Convert back to BGR color space
    result = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

    # Adjust contrast
    result = np.clip(result * contrast_factor, 0, 255).astype(np.uint8)

    return result
```

```python
saturation_factor = 1.5
brightness_factor = 0.5
contrast_factor = 1.5
```

```python
adjusted_image = adjust_saturation_brightness_contrast(im, saturation_factor, brightness_factor, contrast_factor)
```
