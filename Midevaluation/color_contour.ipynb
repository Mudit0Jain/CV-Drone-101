```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def draw_contours_by_color(image_path, color):
    # Read the image
    if color == 'red':
      target_color_lower = [0, 100, 100]
      target_color_upper = [10, 255, 255]
    elif color == 'green':
      target_color_lower = [40, 40, 40]  # Lower bound for green in HSV
      target_color_upper = [80, 255, 255]
    elif color == 'blue':
      target_color_lower = [100, 50, 50]  # Lower bound for blue in HSV
      target_color_upper = [120, 255, 255]
    image = cv2.imread(image_path)

    # Convert the image from BGR to HSV color space
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define the lower and upper bounds for the target color in HSV
    lower_bound = np.array(target_color_lower, dtype=np.uint8)
    upper_bound = np.array(target_color_upper, dtype=np.uint8)

    # Create a binary mask for pixels within the specified color range
    mask = cv2.inRange(hsv_image, lower_bound, upper_bound)

    # Find contours in the binary mask
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on a blank image
    contour_image = image
    cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)
    cv2_imshow(contour_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

image_path = '/content/drive/MyDrive/phots for utility/unzipped_photo1_photo1phone.jpg'


draw_contours_by_color(image_path, 'red')

```
