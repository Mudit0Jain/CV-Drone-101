```python
def imageTransform(s):
  x = np.array(Image.open(s))
  t = x[:,:,0]/999 #+ x[:,:,1]/9
  t = t.astype(float)
  t = np.fft.fft2(t)
  t = np.abs(t)
  return plt.imshow(np.log1p(t), cmap = 'gray')
```
